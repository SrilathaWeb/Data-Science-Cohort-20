## 1. Customers not in the US

```python
#Provide a query showing Customers (just their full names, customer ID and country) who are not in the US.
%%script sqlite3 --column --header chinook.db
SELECT
    FirstName || ' ' || LastName AS FullName,
    CustomerId,
    Country
FROM Customers
WHERE Country != 'USA'
```

## 2. Customers from Brazil


```python
#Provide a query only showing the Customers from Brazil.
%%script sqlite3 --column --header chinook.db
SELECT
    FirstName || ' ' || LastName AS FullName,
    CustomerId,
    Country
FROM Customers
WHERE Country = 'Brazil'
```

## 3. Invoices of customers from Brazil


```python
#Provide a query showing the Invoices of customers who are from Brazil. The resultant table should show the customer's full name, Invoice ID, Date of the invoice and billing country.
%%script sqlite3 --column --header chinook.db
SELECT
    c.FirstName || ' ' || c.LastName AS CustomerName,
    i.InvoiceId,
    i.InvoiceDate,
    i.BillingCountry
FROM Customers c
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
```

## 4. Employees who are Sales Agents

```python
#Provide a query showing only the Employees who are Sales Agents.
%%script sqlite3 --column --header chinook.db
SELECT
    EmployeeId,
    FirstName || ' ' || LastName AS EmployeeName,
    Title
FROM Employees
WHERE Title = 'Sales Support Agent'
```

## 5. Unique list of billing countries from Invoice table

```python
#Provide a query showing a unique list of billing countries from the Invoice table.
%%script sqlite3 --column --header chinook.db
SELECT DISTINCT
    BillingCountry
FROM Invoices
ORDER BY BillingCountry
```

## 6. Invoices of customers from Brazil

```python
#Provide a query showing the invoices of customers who are from Brazil.
%%script sqlite3 --column --header chinook.db
SELECT
    i.*
FROM Invoices i
JOIN Customers c
    ON i.CustomerId = c.CustomerId
WHERE c.Country = 'Brazil'
```

## 7. Invoices associated with each sales agent



```python
#Provide a query that shows the invoices associated with each sales agent. The resultant table should include the Sales Agent's full name.
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    i.InvoiceId,
    i.InvoiceDate,
    i.Total
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
ORDER BY SalesAgentName, i.InvoiceId
```

## 8. Invoice Total, Customer name, Country and Sales Agent name for all invoices

```python
#Provide a query that shows the Invoice Total, Customer name, Country and Sale Agent name for all invoices and customers.
%%script sqlite3 --column --header chinook.db
SELECT
    i.Total AS InvoiceTotal,
    c.FirstName || ' ' || c.LastName AS CustomerName,
    c.Country,
    e.FirstName || ' ' || e.LastName AS SalesAgentName
FROM Invoices i
JOIN Customers c
    ON i.CustomerId = c.CustomerId
JOIN Employees e
    ON c.SupportRepId = e.EmployeeId
ORDER BY i.InvoiceId
```

## 9. Number of invoices and total sales in 2009 and 2011

```python
#How many Invoices were there in 2009 and 2011? What are the respective total sales for each of those years?
%%script sqlite3 --column --header chinook.db
SELECT
    strftime('%Y', InvoiceDate) AS InvoiceYear,
    COUNT(*) AS NumberOfInvoices,
    ROUND(SUM(Total), 2) AS TotalSales
FROM Invoices
WHERE strftime('%Y', InvoiceDate) IN ('2009', '2011')
GROUP BY strftime('%Y', InvoiceDate)
ORDER BY InvoiceYear
```

## 10. Count line items for Invoice ID 37

```python
#Looking at the InvoiceLine table, provide a query that COUNTs the number of line items for Invoice ID 37.
%%script sqlite3 --column --header chinook.db
SELECT
    InvoiceId,
    COUNT(*) AS LineItemCount
FROM Invoice_Items
WHERE InvoiceId = 37
GROUP BY InvoiceId
```

## 11. Count line items for each Invoice

```python
#Looking at the InvoiceLine table, provide a query that COUNTs the number of line items for each Invoice. HINT: GROUP BY
%%script sqlite3 --column --header chinook.db
SELECT
    InvoiceId,
    COUNT(*) AS LineItemCount
FROM Invoice_Items
GROUP BY InvoiceId
ORDER BY InvoiceId
```

## 12. Track name with each invoice line item

```python
#Provide a query that includes the track name with each invoice line item.
%%script sqlite3 --column --header chinook.db
SELECT
    il.InvoiceLineId,
    il.InvoiceId,
    t.Name AS TrackName,
    il.UnitPrice,
    il.Quantity
FROM Invoice_Items il
JOIN Tracks t
    ON il.TrackId = t.TrackId
ORDER BY il.InvoiceLineId
```

## 13. Purchased track name and artist name with each invoice line item

```python
#Provide a query that includes the purchased track name AND artist name with each invoice line item.
%%script sqlite3 --column --header chinook.db
SELECT
    il.InvoiceLineId,
    il.InvoiceId,
    t.Name AS TrackName,
    ar.Name AS ArtistName,
    il.UnitPrice,
    il.Quantity
FROM Invoice_Items il
JOIN Tracks t
    ON il.TrackId = t.TrackId
JOIN Albums al
    ON t.AlbumId = al.AlbumId
JOIN Artists ar
    ON al.ArtistId = ar.ArtistId
ORDER BY il.InvoiceLineId
```

## 14. Number of invoices per country

```python
#Provide a query that shows the # of invoices per country. HINT: GROUP BY
%%script sqlite3 --column --header chinook.db
SELECT
    BillingCountry,
    COUNT(*) AS NumberOfInvoices
FROM Invoices
GROUP BY BillingCountry
ORDER BY NumberOfInvoices DESC
```

## 15. Total number of tracks in each playlist

```python
#Provide a query that shows the total number of tracks in each playlist. The Playlist name should be include on the resultant table.
%%script sqlite3 --column --header chinook.db
SELECT
    p.Name AS PlaylistName,
    COUNT(pt.TrackId) AS NumberOfTracks
FROM Playlists p
JOIN Playlist_Track pt
    ON p.PlaylistId = pt.PlaylistId
GROUP BY p.PlaylistId, p.Name
ORDER BY NumberOfTracks DESC
```

## 16. All Tracks with no IDs, including Album, Media Type and Genre

```python
#Provide a query that shows all the Tracks, but displays no IDs. The resultant table should include the Album name, Media type and Genre.
%%script sqlite3 --column --header chinook.db
SELECT
    t.Name AS TrackName,
    al.Title AS AlbumName,
    mt.Name AS MediaType,
    g.Name AS Genre,
    t.Composer,
    t.Milliseconds,
    t.Bytes,
    t.UnitPrice
FROM Tracks t
JOIN Albums al
    ON t.AlbumId = al.AlbumId
JOIN Media_Types mt
    ON t.MediaTypeId = mt.MediaTypeId
JOIN Genres g
    ON t.GenreId = g.GenreId
ORDER BY t.Name
```

## 17. All Invoices including number of invoice line items

```python
#Provide a query that shows all Invoices but includes the # of invoice line items.
%%script sqlite3 --column --header chinook.db
SELECT
    i.InvoiceId,
    i.CustomerId,
    i.InvoiceDate,
    i.BillingAddress,
    i.BillingCity,
    i.BillingState,
    i.BillingCountry,
    i.BillingPostalCode,
    i.Total,
    COUNT(il.InvoiceLineId) AS NumberOfLineItems
FROM Invoices i
JOIN Invoice_Items il
    ON i.InvoiceId = il.InvoiceId
GROUP BY
    i.InvoiceId,
    i.CustomerId,
    i.InvoiceDate,
    i.BillingAddress,
    i.BillingCity,
    i.BillingState,
    i.BillingCountry,
    i.BillingPostalCode,
    i.Total
ORDER BY i.InvoiceId
```

## 18. Total sales made by each sales agent

```python
#Provide a query that shows total sales made by each sales agent.
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    ROUND(SUM(i.Total), 2) AS TotalSales
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
GROUP BY e.EmployeeId, SalesAgentName
ORDER BY TotalSales DESC
```

## 19. Sales agent with the most sales in 2009

```python
#Which sales agent made the most in sales in 2009?
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    ROUND(SUM(i.Total), 2) AS TotalSales
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
  AND strftime('%Y', i.InvoiceDate) = '2009'
GROUP BY e.EmployeeId, SalesAgentName
ORDER BY TotalSales DESC
LIMIT 1
```

## 20. Sales agent with the most sales in 2010

```python
#Which sales agent made the most in sales in 2010?
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    ROUND(SUM(i.Total), 2) AS TotalSales
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
  AND strftime('%Y', i.InvoiceDate) = '2010'
GROUP BY e.EmployeeId, SalesAgentName
ORDER BY TotalSales DESC
LIMIT 1
```

## 21. Sales agent with the most sales overall

```python
#Which sales agent made the most in sales over all?
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    ROUND(SUM(i.Total), 2) AS TotalSales
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
JOIN Invoices i
    ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
GROUP BY e.EmployeeId, SalesAgentName
ORDER BY TotalSales DESC
LIMIT 1
```

## 22. Number of customers assigned to each sales agent

```python
#Provide a query that shows the # of customers assigned to each sales agent.
%%script sqlite3 --column --header chinook.db
SELECT
    e.FirstName || ' ' || e.LastName AS SalesAgentName,
    COUNT(c.CustomerId) AS NumberOfCustomers
FROM Employees e
JOIN Customers c
    ON e.EmployeeId = c.SupportRepId
WHERE e.Title = 'Sales Support Agent'
GROUP BY e.EmployeeId, SalesAgentName
ORDER BY NumberOfCustomers DESC
```

## 23. Total sales per country and highest spending country

```python
#Provide a query that shows the total sales per country. Which country's customers spent the most?
%%script sqlite3 --column --header chinook.db
SELECT
    BillingCountry,
    ROUND(SUM(Total), 2) AS TotalSales
FROM Invoices
GROUP BY BillingCountry
ORDER BY TotalSales DESC
```

## 24. Most purchased track of 2013

```python
#Provide a query that shows the most purchased track of 2013.
%%script sqlite3 --column --header chinook.db
SELECT
    t.Name AS TrackName,
    SUM(il.Quantity) AS TotalPurchased
FROM Invoice_Items il
JOIN Invoices i
    ON il.InvoiceId = i.InvoiceId
JOIN Tracks t
    ON il.TrackId = t.TrackId
WHERE strftime('%Y', i.InvoiceDate) = '2013'
GROUP BY t.TrackId, t.Name
ORDER BY TotalPurchased DESC
LIMIT 1
```

## 25. Top 5 most purchased tracks overall

```python
#Provide a query that shows the top 5 most purchased tracks over all.
%%script sqlite3 --column --header chinook.db
SELECT
    t.Name AS TrackName,
    SUM(il.Quantity) AS TotalPurchased
FROM Invoice_Items il
JOIN Tracks t
    ON il.TrackId = t.TrackId
GROUP BY t.TrackId, t.Name
ORDER BY TotalPurchased DESC
LIMIT 5
```

## 26. Top 3 best selling artists


```python
#Provide a query that shows the top 3 best selling artists.
%%script sqlite3 --column --header chinook.db
SELECT
    ar.Name AS ArtistName,
    ROUND(SUM(il.UnitPrice * il.Quantity), 2) AS TotalSales
FROM Invoice_Items il
JOIN Tracks t
    ON il.TrackId = t.TrackId
JOIN Albums al
    ON t.AlbumId = al.AlbumId
JOIN Artists ar
    ON al.ArtistId = ar.ArtistId
GROUP BY ar.ArtistId, ar.Name
ORDER BY TotalSales DESC
LIMIT 3
```

## 27. Most purchased Media Type

```python
#Provide a query that shows the most purchased Media Type.
%%script sqlite3 --column --header chinook.db
SELECT
    mt.Name AS MediaType,
    SUM(il.Quantity) AS TotalPurchased
FROM Invoice_Items il
JOIN Tracks t
    ON il.TrackId = t.TrackId
JOIN Media_Types mt
    ON t.MediaTypeId = mt.MediaTypeId
GROUP BY mt.MediaTypeId, mt.Name
ORDER BY TotalPurchased DESC
LIMIT 1
```
